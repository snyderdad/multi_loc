```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import scipy.fftpack as fft
```

```python
%matplotlib inline
```

```python
n = 200
Cx = np.arange(n)
Cx, Cy = np.meshgrid(Cx, Cx)

rho = Cx.copy()*np.nan
for i in range(n):
    for j in range(n):
        rho[i,j]=min(abs(i-j),(n-abs(i-j))%n)

plt.figure()
im = plt.imshow(rho)
plt.colorbar(im)
```

```python
l1 = 2
C1 = np.exp(-rho**2/(2*l1**2))
l2 = 30
C2 = np.exp(-rho**2/(2*l2**2))
vmin=0
vmax=1
plt.figure()
im = plt.imshow(C1, vmin=vmin, vmax=vmax)
plt.colorbar(im)

plt.figure()
im = plt.imshow(C2, vmin=vmin, vmax=vmax)
plt.colorbar(im)

C3 = (C1 + C2)/2
plt.figure()
im = plt.imshow(C3, vmin=vmin, vmax=vmax)
plt.colorbar(im)
```

```python
this = fft.fft2(C3)
this = fft.fftshift(this)
plt.figure()
im = plt.imshow(this.real)
plt.colorbar(im)

plt.figure()
im = plt.imshow(np.log(1 + np.abs(this.real)))
plt.colorbar(im)

plt.figure()
im = plt.imshow(this.imag)
plt.colorbar(im)

plt.figure()
im = plt.imshow(np.log(1 + np.abs(this.imag)))
plt.colorbar(im)

mag = np.sqrt(this.real**2 + this.imag**2)

plt.figure()
im = plt.imshow(mag)
plt.colorbar(im)

plt.figure()
im = plt.imshow(np.log(1 + np.abs(mag)))
plt.colorbar(im)
```
